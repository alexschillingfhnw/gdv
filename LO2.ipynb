# LO2: Visual Perception
For the second learning objective, I dived into the world of visual perception and I learned how to apply perception for designs. The dataset I used for this chapter was also found on Kaggle: https://www.kaggle.com/datasets/tanuprabhu/population-by-country-2020

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_penguins = pd.read_csv("penguins.csv")
df_penguins.head()
```

```python
plt.figure(figsize=(7,7))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)"
)
plt.title("Culmen Length vs Flipper Length")
plt.show()
```

## Visual Variables
A visual variable in data visualization is an aspect of a graphical object that can visually differentiate it from other objects and can be controlled during the design process. This concept was created by Bertin, who identified seven main categories of visual variables: position, size, shape, value, color, orientation, and texture. („Visual Variable“, 2022)
### Orientation

```python
# Switch x & y 

plt.figure(figsize=(7,7))

sns.scatterplot(
    data = df_penguins,
    y = "Culmen Length (mm)", 
    x = "Flipper Length (mm)"
)
plt.title("Flipper Length vs Culmen Length")
plt.show()
```

### Size

```python
# Change size of datapoints

plt.figure(figsize=(7,7))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)",
    size = "Body Mass (g)"
)

plt.title("Culmen Length vs Flipper Length by Body Mass")
plt.show()
```

### Hue

```python
# Add 3rd variable species to change hue of data points

plt.figure(figsize=(7,7))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)",
    hue = "Species"
)
plt.title("Culmen Length vs Flipper Length by Species")
plt.show()
```

```python
plt.figure(figsize=(7,7))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)", 
    hue = "Sex"
)
plt.title("Culmen Length vs Flipper Length by Sex")
plt.show()
```

### Shape

```python
plt.figure(figsize=(7,7))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)",
    marker = "s"
)
plt.title("Culmen Length vs Flipper Length by Sex")
plt.show()
```

```python
plt.figure(figsize=(7,7))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)",
    style = "Sex"
)
plt.title("Culmen Length vs Flipper Length by Sex")
plt.show()
```

### Transparency

```python
plt.figure(figsize=(7,7))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)", 
    alpha = 0.4
)
plt.title("Culmen Length vs Flipper Length with adjusted transparency")
plt.show()
```

```python
plt.figure(figsize=(7,7))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)",
    color = "Red",
    alpha = 0.3
)
plt.title("Culmen Length vs Flipper Length")
plt.show()
```
