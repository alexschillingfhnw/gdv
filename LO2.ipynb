```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_china = pd.read_csv("china_population.csv")
df_china
```

```python
df_penguins = pd.read_csv("penguins.csv", index_col='Unnamed: 0')
df_penguins.head()
```

## Visual Variables
A visual variable in data visualization is an aspect of a graphical object that can visually differentiate it from other objects and can be controlled during the design process. This concept was created by Bertin, who identified seven main categories of visual variables: position, size, shape, value, color, orientation, and texture. (âVisual Variableâ, 2022)
### Size

```python
plt.figure(figsize=(9,5))
sns.scatterplot(x="Year", y="Population", size="Fertility Rate", sizes=(10, 200), data=df_china)
plt.title("Year vs Population")
plt.show()

plt.figure(figsize=(9,5))
sns.scatterplot(x="Year", y="Urban Population", size="Urban Pop %", sizes=(10, 100), data=df_china)
plt.title("Year vs Urban Population")
plt.show()
```

```python
plt.figure(figsize=(8,7))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)", 
    size = "Body Mass (g)",
)

plt.show()
```

### Hue

```python
plt.figure(figsize=(9,5))

sns.scatterplot(
    data=df_china,
    x="Year", 
    y="Population", 
    hue="Fertility Rate",
)

plt.title("Scatterplot: Year vs Population")
plt.show()
```

```python
plt.figure(figsize=(7,6))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)",
    hue = "Species", 
    palette ="Set2"
)

plt.show()
```

```python
plt.figure(figsize=(7,6))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)", 
    hue = "Sex",
    palette = "Set2"
)
plt.show()
```

### Shape

```python
plt.figure(figsize=(7,6))

sns.scatterplot(
    data = df_penguins,
    x = "Culmen Length (mm)", 
    y = "Flipper Length (mm)", 
    style = "Sex"
)

plt.show()
```
