```python
data = pd.read_csv('EPL_20_21.csv')
data
```

### 20/21 Premier League Goals by Club

I wanted to display the amount of goals scored by each club in the 20/21 premier league season. To achieve this, I needed to prepare the data. Firstly, I grouped the dataset by club. After that I selected the 'Club' and 'Goals' columns and summed the goals scored by each club. Next, the dataset was sorted by the total scored goals in ascending order.

```python
club_goals = data.groupby('Club')[['Club', 'Goals']].sum()

club_goals_sorted = club_goals.sort_values('Goals', ascending=True).reset_index()
club_goals_sorted
```

Moving on to the visualization, a flipped bar chart (where the x and y axes are swapped) is ideal to visualize the amount of goals scored by each club. After swapping the axes, a compact figure in which all visual elements, including all text, are horizontally oriented  is obtained, which makes the figure easy to read.

With the 'scale_x_discrete' function, the clubs (bars) are sorted by goals. This type of rearragement is possible, because there is no natural ordering to the clubs.

### Bar Chart?

```python
(ggplot(club_goals_sorted, aes(x=club_goals_sorted.Club, y=club_goals_sorted.Goals))
 + geom_col(fill="#509FF0", width=.5)
 + coord_flip()
 + scale_x_discrete(limits=club_goals_sorted.Club)
 + labs(y='Goals Scored', x='Club', title='20/21 Premier League Goals by Club')
)
```

### Dot Plot?

```python
(ggplot(club_goals_sorted, aes(x=club_goals_sorted.Club, y=club_goals_sorted.Goals))
 + geom_point(color="#509FF0", size=2)
 + coord_flip()
 + scale_x_discrete(limits=club_goals_sorted.Club)
 + labs(y='Goals Scored', x='Club', title='20/21 Premier League Goals by Club')
)
```

### 20/21 Premier League Matches by Age

Next I wanted to figure out, which age groups play the most matches.

Unlike the last visualization, this one has a natural ordering of the categorial variable 'Age'. In this case, the bars were arranged in order of the increasing age.

```python
(ggplot(data, aes(x=data.Age, y=data.Matches))
 + geom_col(fill="#509FF0", width=0.9)
 + labs(y='Goals Scored', x='Age', title='20/21 Premier League Matches by Age')
)
```

### 20/21 Premier League Goals by position

Next I would like to visualize, which positions score the most goals.
- GK = Goalkeeper
- DF = Defender
- MF = Midfielder
- FW = Forward

```python
position_goals = data.groupby('Position')[['Position', 'Goals']].sum()

position_goals_sorted = position_goals.sort_values('Goals', ascending=False).reset_index()

position_goals_sorted
```

```python
(ggplot(position_goals_sorted, aes(x=position_goals_sorted.Position, y=position_goals_sorted.Goals))
 + geom_col(fill="#509FF0", width=0.9)
 #+ coord_flip()
 + scale_x_discrete(limits=position_goals_sorted.Position)
 + labs(y='Goals Scored', x='Position', title='20/21 Premier League Goals by position')
)
```

### Passes Attempted vs Succuessfull Passes (%) by Position

Another interesting visualization is to see the amount of attempted and completed passes by each player and their position. What is diaplayed in the plot corresponds with reality. The most passes are attempted by midfielders and defenders, because they bring the ball forward to create chances for the forwards. The forwards attempt less passes on average, as they are the players making the runs into dangerous spaces to receive the ball from positions behind them. What's also visible is that Goalkeepers have the lowest completed passing rate, because most of the time they launch the ball forward & cannot builup play.

Football theory aside, the visualization accurately conveys and does not mislead the data. 

```python
passes = data.drop(data[data.Passes_Attempted < 1].index)
passes = data.drop(data[data.Perc_Passes_Completed < 1].index)

(ggplot(passes, aes(x='Perc_Passes_Completed', y='Passes_Attempted', color='Position'))
 + geom_point(size = 0.7)
 + labs(y='Passes Attempted', x='Successfull Passes (%)', title='Passes Attempted vs Succuessfull Passes (%) by Position')
)
```
